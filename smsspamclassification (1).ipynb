```python
def currency(data):
    currency_symbols=['$', '£', '€', '¥','₹']
    for i in currency_symbols:
        if i in data:
            return 1
    return 0
    
```

```python
dataset['contains_currency_cymbols']=dataset['message'].apply(currency)
```

```python
dataset
```

```python
plt.figure(figsize=(8,8))
g=sns.countplot(x='contains_currency_cymbols',data=dataset,hue='labels')
p=plt.title('countplot for containing currency symbol')
p=plt.xlabel('does the sms contain any currency symbol')
p=plt.ylabel('count')
p=plt.legend(labels=['ham','spam'],loc=9)
           
```

```python
def number(data):
    for i in data:
        if ord(i) >=48 and ord(i) <=57:
            return 1
    return 0
```

```python
dataset['contains_numbers']=dataset['message'].apply(number)
```

```python
import nltk
import re
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
corpus=[]
wnl=WordNetLemmatizer()
for sms in list(dataset.message):
    message = re.sub(pattern='[^a-zA-Z]', repl=' ', string=sms)
    message = message.lower()
    words= message.split()
    filtered_words = [word for word in words if word not in set(stopwords.words('english'))]
    lenn_words = [wnl.lemmatize(word) for word in filtered_words]
    message=' '.join(lenn_words)
    corpus.append(message)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=500)
vectors=tfidf.fit_transform(corpus).toarray()
feature_names=tfidf.get_feature_names_out()
```

```python
x = pd.DataFrame(vectors, columns=feature_names)
y=dataset['labels']
```

```python
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
cv=cross_val_score(mnb,x,y,scoring='f1',cv=10)
print(round(cv.mean(),3))
print(round(cv.std(),3))
```

```python
mnb.fit(x_train,y_train)
y_pred=mnb.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(8,8))
axis_labels=['ham','spam']
g=sns.heatmap(data=cm, xticklabels=axis_labels, yticklabels=axis_labels, annot=True, fmt='g', cbar_kws={'shrink': 0.5})
p=plt.title('confusion matrix of multinomial bayes model')
p=plt.xlabel('actual values')
p=plt.ylabel('predicted values')             
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
cv1=cross_val_score(dt,x,y,scoring='f1',cv=10)
print(round(cv1.mean(),3))
print(round(cv1.std(),3))
```

```python
dt.fit(x_train,y_train)
y_pred1=dt.predict(x_test)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
cm=confusion_matrix(y_test,y_pred1)
cm
```

```python
plt.figure(figsize=(8,8))
axis_labels=['ham','spam']
g=sns.heatmap(data=cm, xticklabels=axis_labels, yticklabels=axis_labels, annot=True, fmt='g', cbar_kws={'shrink': 0.5})
p=plt.title('confusion matrix of multinomial bayes model')
p=plt.xlabel('actual values')
p=plt.ylabel('predicted values')  
```

```python
def predict_spam(sms):
    message = re.sub(pattern='[^a-zA-Z]', repl=' ', string=sms)
    message = message.lower()
    words= message.split()
    filtered_words = [word for word in words if word not in set(stopwords.words('english'))]
    lenn_words = [wnl.lemmatize(word) for word in filtered_words]
    message=' '.join(lenn_words)
    temp=tfidf.transform([message]).toarray()
    return dt.predict(temp)
```
